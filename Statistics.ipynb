```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from scipy.stats import binom #library for binomial distribution calculations.
from scipy.stats import poisson
```

```python
df = pd.read_csv('/content/drive/MyDrive/CA MSc/Estimated Migration.csv')

df.head(5)
```

```python
# ploting price distribution
plt.figure(figsize=(10,4))
sns.distplot(df.VALUE)

plt.show()
```

```python
# Outlier Analysis
sns.boxplot(df['VALUE'])
plt.show()
```

```python
age_group_data = df.groupby('Age Group')['VALUE'].sum()
age_group_data.plot(kind='bar')
plt.xlabel('Age Group')
plt.ylabel('Total Migration (Thousands)')
plt.title('Migration by Age Group')
plt.show()
```

```python
sex_data = df.groupby('Sex')['VALUE'].sum()
sex_data.plot(kind='bar')
plt.xlabel('Sex')
plt.ylabel('Total Migration (Thousands)')
plt.title('Migration by Sex')
plt.show()
```

```python
# 4. Yearly Trends
yearly_data = df.groupby('Year')['VALUE'].sum()
plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_data.index, y=yearly_data.values)
plt.xlabel('Year')
plt.ylabel('Total Net Migration (Thousand)')
plt.title('Yearly Trends in Net Migration')
plt.show()
```

```python
#Binomial Distribution
```

```python
# Parameters
n = 20  # Total number of immigration applications
p = 0.30  # Probability of an application being approved
```

```python
# Calculate the probability of success for different numbers of approved applications
k_values = np.arange(0, n + 1)
probabilities = binom.pmf(k_values, n, p)
```

```python
# Create a bar plot
plt.bar(k_values, probabilities, align='center', alpha=0.7)
plt.xlabel('Number of Approved Applications')
plt.ylabel('Probability')
plt.title('Binomial Distribution for Immigration Success')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#Poisson
```

```python
# Parameters
n = 20  # Total number of immigration applications
p = 0.30  # Probability of an application being approved

# Calculate the rate parameter for the Poisson distribution
lambda_poisson = n * p
```

```python
# Calculate the Poisson probabilities for different numbers of approved applications
k_values = np.arange(0, n + 1)
poisson_probabilities = poisson.pmf(k_values, lambda_poisson)
```

```python
# Create a plot to visualize the Poisson distribution
plt.figure(figsize=(8, 4))
plt.bar(k_values, poisson_probabilities, align='center', alpha=0.6, color='b')
plt.xlabel('Number of Approved Applications')
plt.ylabel('Probability')
plt.title('Poisson Distribution for Approved Applications')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#Normal distribution
```

```python
from scipy.stats import poisson, binom, norm
```

```python
# Calculate the Normal distribution using mean and standard deviation
mean_normal = n * p
std_dev_normal = np.sqrt(n * p * (1 - p))
x_values = np.linspace(0, n, 1000)
normal_probabilities = norm.pdf(x_values, mean_normal, std_dev_normal)
```

```python
# Create a range of x values for the Normal distribution
x_values = np.linspace(0, n, 1000)
```

```python
# Calculate the Normal distribution probabilities for each x value
normal_probabilities = norm.pdf(x_values, mean_normal, std_dev_normal)
```

```python
# Create a plot to visualize the Normal distribution
plt.figure(figsize=(8, 4))
plt.plot(x_values, normal_probabilities, label='Normal', linestyle='-', color='r')
plt.xlabel('Number of Approved Applications')
plt.ylabel('Probability Density')
plt.title('Normal Distribution for Approved Applications')
plt.axvline(mean_normal, color='red', linestyle='--', label='Mean')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Calculate the Binomial probabilities
k_values = np.arange(0, n + 1)
binom_probabilities = binom.pmf(k_values, n, p)
```

```python
# Calculate the Binomial probabilities for each value of k
binom_probabilities = binom.pmf(k_values, n, p)
```

```python
# Print the probabilities
for k, probability in zip(k_values, binom_probabilities):
    print(f'P(X = {k}) = {probability:.4f}')
```

```python
# Create a plot to compare the three distributions
plt.plot(k_values, binom_probabilities, label='Binomial', marker='o', linestyle='-')
plt.plot(k_values, poisson_probabilities, label='Poisson', marker='s', linestyle='--')
plt.plot(x_values, normal_probabilities, label='Normal', linestyle='-')
plt.xlabel('Number of Approved Applications')
plt.ylabel('Probability')
plt.title('Comparison of Poisson, Binomial, and Normal Distributions')
plt.xticks(k_values)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
